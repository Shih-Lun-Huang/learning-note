# 介紹

我們將從機器學習模型如何工作以及大概如何使用開始。如果之前進行過統計建模或機器學習，則可能會覺得很基礎。不用擔心，我們將盡快構建強大的模型。

通過此微型課程，您可以在完成以下場景時構建模型：

你的堂兄已經賺了數百萬美元，投資房地產。由於您對數據科學的興趣，他願意與您成為業務合作夥伴。他將提供資金，並且您將提供預測各種房價的模型。

你問表弟，他過去如何預測房地產價值。他說這只是直覺。但是更多的質疑顯示他從過去的房屋中識別出了價格模式，然後他利用這些模式對他正在考慮的新房屋做出預測。

機器學習的工作方式相同。我們將從稱為決策樹(Decision Tree)的模型開始。有更高級的模型可以提供更準確的預測，但是決策樹比較容易理解，它們是數據科學中一些最佳模型的基本觀念。

為簡單起見，我們將從最簡單的決策樹開始。

```python
from IPython.display import Image 
```

```python
Image(filename = "img/L2-1.png", width = 500, height = 300)
```

它將房屋僅分為兩類。 所有房屋預測價格的考量為:同一類別房屋的歷史平均價格。

我們使用數據來決定如何將房屋分為兩組，然後再次確定每組中的預測價格。 從數據捕獲模式的這一步驟稱為**擬合(fitting)**或**訓練(training)**模型。 用於**擬合**模型的數據稱為**訓練數據**。

有關模型擬合方式的詳細信息（例如，如何拆分數據）非常複雜，我們留著以後會提到。 擬合模型後，您可以將其應用於新數據以**預測(predict)**其他房屋的價格。

左側的決策樹（決策樹1）可能更有意義，因為它反映了現實情況，即臥室多的房屋比臥室少的房屋的價格更高。 該模型的最大缺點是，它沒有涵蓋影響房價的大多數因素，例如浴室數量，地段大小，位置等。

我們可以使用具有更多“分裂”的樹來捕獲更多因素。 這些被稱為“更深”的樹。 還會考慮每棟房屋總面積的決策樹可能如下所示：

```python
Image(filename = "img/L2-3.png", width = 500, height = 400)
```

我們可以通過追蹤決策樹來預測任何房屋的價格，並始終選擇與該房屋的特徵相對應的路徑。 房屋的估計價格在樹的底部。我們進行預測的底部點稱為葉(leaf)。

葉上的拆分和值將由數據確定，因此是時候檢查要使用的數據了。
