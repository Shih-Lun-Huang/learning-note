# 解釋數據描述

結果顯示原始數據集中每列有8個數字。 第一個數字（**count計數**）顯示**有多少行具有非缺失值**。

缺失值的原因很多。例如，在測量一間臥室的房子時，不會收集第二間臥室的大小。<br>我們將回到缺少數據的主題。

第二個值是**mean**，即平均值。 在此之下，**std**是標準偏差，它測量數值在數值上的**分散程度**。

要解釋**min(最小值)**，**25％**，**50％**，**75％**和**max(最大值)**，請想像從最低值到最高值對每列進行排序。 第一個（最小）值是 **min**。 如果您瀏覽列表的四分之一，則會發現一個數字，該數字大於值的25％，小於值的75％。 那是**25％**的值（讀作“ 25th percentile 25個百分位數”）。 第50個百分位數和第75個百分位數的定義類似，**max**是最大數。
